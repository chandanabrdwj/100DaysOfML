```python
import pandas as pd
```

```python
df_country = pd.read_csv('../input/best-cities-and-countries-for-startups-in-2021/Best Countries for Startups.csv')
df_city = pd.read_csv('../input/best-cities-and-countries-for-startups-in-2021/Best Cities for Startups.csv')
```

```python
#Strip trailing whitespaces and replace space with underscore

df_country.columns = df_country.columns.str.strip()
df_country.columns = df_country.columns.str.replace(' ','_')

df_city.columns = df_city.columns.str.strip()
df_city.columns = df_city.columns.str.replace(' ','_')
```

```python
# Top 10 ranked
df_country.head(10)
```

```python
# Top 10 positive change in position
df_country[df_country.change_in_position_sign == '+'].head(10)
```

```python
# Top 10 negative change in position
df_country[df_country.change_in_position_sign == '-'].head(10)
```

```python
# Countries with quality score greater than 10
df_country[df_country.quality_score > 10]
```

```python
# countries with quality score greater than 5
df_country[df_country.quality_score > 5]
```

```python
# Countries with quality score 3 and above that are growing positively
df_country[(df_country.quality_score > 3) & (df_country.change_in_position_sign == '+')]
```

```python
# Countries with quality score above 3 that are on the decline
df_country[(df_country.quality_score > 3) & (df_country.change_in_position_sign == '-')]
```
