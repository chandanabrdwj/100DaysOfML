```python
import pandas as pd
```

```python
Country = pd.read_csv('../input/best-cities-and-countries-for-startups-in-2021/Best Countries for Startups.csv')
City = pd.read_csv('../input/best-cities-and-countries-for-startups-in-2021/Best Cities for Startups.csv')
```

```python
#Strip trailing whitespaces and replace space with underscore

Country.columns = Country.columns.str.strip()
Country.columns = Country.columns.str.replace(' ','_')

City.columns = City.columns.str.strip()
City.columns = City.columns.str.replace(' ','_')
```

```python
# Top 10 ranked
Country[Country.ranking < 11]
```

```python
# Top 10 negative change in position
Country[Country.change_in_position_sign == '-'].head(10)
```

```python
# Countries with quality score greater than 10
Country[Country.quality_score > 10]
```

```python
# countries with quality score greater than 5
Country[Country.quality_score > 5]
```

```python
# Countries with quality score 3 and above that are growing positively
Country[(Country.quality_score > 3) & (Country.change_in_position_sign == '+')]
```

```python
# Countries with quality score above 3 that are on the decline
Country[(Country.quality_score > 3) & (Country.change_in_position_sign == '-')]
```
